# Demographic Overview

In this notebook, we'll get an overview of the demographic distribution of the sample. 

Starting off, we'll import all relevant libraries for this describtive part of the analysis and load the preprocessed dataset by using the load_and_preprocess function specified in the corresponding document. 

```python
import matplotlib.pyplot as plt

from data_preprocessing import load_and_preprocess

# Load and preprocess data
df = load_and_preprocess('data/cleaned_file.csv')
```

## Visualizing the Gender Distribution:

```python
# Numerical Gender distribution
print(df['gender'].value_counts())

# Gender bar plot
df['gender'].value_counts().plot(kind='bar', title='Gender Distribution')
plt.ylabel('Count')
plt.show()
```

## Visualizing the Age Distribution:

```python
# Numeric Age distribution
print(df['age'].value_counts().sort_index())

# Age bar plot
df['age'].value_counts().sort_index().plot(kind='bar', title='Age Distribution')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Visualizing the Education Level Distribution:

```python
# Education level Distribution
print(df['education_level'].value_counts())

# Education level plot
df['education_level'].value_counts().plot(kind='bar', title='Education Level Distribution')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Visualizing the Voting Experience Distribution: 

```python
# Voting experience
print(df['voting_experience'].value_counts())

# Voting experience plot
df['voting_experience'].value_counts().plot(kind='bar', title='Voting Experience')
plt.ylabel('Count')
plt.show()
```
